```python
for industry in invest_data:
    list_orders = []
    list_strings = []
    for company in invest_data[industry]:
        order = []
        string = ""
        counter = 0
        for entry in invest_data[industry][company]:
            print("order", order)
            print(entry[1] in order)
            if entry[1] in order:
                continue
            if entry[1] == "series_unknown":
                continue
            else:
                order.append(entry[1])
                if string == "":
                    string = entry[1]
                    counter += 1
                else:
                    if counter == 6:
                        break
                    else:
                        string += "-" + entry[1]
                        counter += 1
        if (counter !=6 and counter >0):
            string += "-end"
        if counter >0:
            list_orders.append(tuple(order))
            list_strings.append(string)
    strings_dict[industry] = list_strings
```
