```python
import pandas as pd
import numpy as np
from datetime import datetime
import operator
```

```python
companies = pd.read_excel("../V2_cb_Nov18_unicorns.xlsx")
```

```python
type_dict = {
    "series_e": "Series E",
    "series_d": "Series D",
    "private_equity": "Pvt Equity",
    "corporate_round": "Corporate",
    "series_c": "Series C",
    "debt_financing": "Debt",
    "secondary_market": "Secondary",
    "undisclosed": "Undisclosed",
    "series_f": "Series F",
    "series_a": "Series A",
    "series_h": "Series H",
    "series_g": "Series G",
    "angel": "Angel",
    "series_b": "Series B",
    "series_i": "Series I",
    "equity_crowdfunding": "Crowdfunding",
    "product_crowdfunding": "Crowdfunding",
    "seed": "Seed",
    "convertible_note": "Conv Note",
    "post_ipo_equity": "Post IPO Equity",
    "non_equity_assistance": "Assistance",
    "post_ipo_debt": "Post IPO Debt",
    "initial_coin_offering": "ICO",
    "grant": "Grant",
    "series_j": "Series J",
    "series_unknown":  "series_unknown"
}
```

```python
# companies founded in 2010
year = 2000
companies_list = []
for no, row in companies.iterrows():
    if type(row["founded_on"]) != str:
        continue
    print(row["founded_on"])
    if datetime.strptime(row["founded_on"], '%Y-%m-%d').year == year:
        companies_list.append(row["company_name"])
```

```python
len(companies_list)
```

```python
investments = pd.read_excel("../V2_cb_Nov18_unicorns.xlsx", sheet_name="Rounds")
```

```python
investments
```

```python
invest_data = {}
```

```python
for no, row in investments.iterrows():
    company = row["company_name"]
    if company not in companies_list:
        continue
    ftype = type_dict[row["funding_round_type"]]
    date = row["announced_on"]
    if company in invest_data:
        invest_data[company].append((date, ftype))
    else:
        invest_data[company]= []
        invest_data[company].append((date, ftype))
```

```python
invest_data
```

```python
sorted_data={}
for key in invest_data:
    sorted_data[key] = sorted(invest_data[key], key=operator.itemgetter(0))
```

```python
sorted_data
```

```python
for entry in sorted_data:
    sorted_data[entry] = list(set(sorted_data[entry]))
```

```python
sorted_data
```

```python
for key in sorted_data:
    sorted_data[key] = sorted(sorted_data[key], key=operator.itemgetter(0))
```

```python
sorted_data
```

```python
list_orders = []
list_strings = []
for key in sorted_data:
    order = []
    string = ""
    counter = 0
    for entry in sorted_data[key]:
        print("order", order)
        print(entry[1] in order)
        if entry[1] in order:
            continue
        if entry[1] == "series_unknown":
            continue
        else:
            order.append(entry[1])
            if string == "":
                string = entry[1]
                counter += 1
            else:
                if counter == 6:
                    break
                else:
                    string += "-" + entry[1]
                    counter += 1
 
    #if (counter !=6 and counter >0):
    #    string += "-end"
    if counter >0:
        list_orders.append(tuple(order))
        list_strings.append(string)
        
```

```python
list_orders
list_strings
```

```python
dict_numbers = {}
```

```python
for item in list_strings:
    if item in dict_numbers:
        dict_numbers[item] += 1
    else:
        dict_numbers[item] = 1
```

```python
dict_numbers
```

```python
dataframe_1 = pd.DataFrame.from_dict(dict_numbers, orient='index')
```
